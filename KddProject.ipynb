### Data Preparation (what was done to prepare the data)

Steps we followed In this below cell are 

1. **Replace the Infinite values** 
     - We start by replacing infinite values in the DataFrame with `NaN` to handle them later.
2. **Handle Missing Values**
    - For numeric columns, we fill any `NaN` values with the median of each respective column.
3. **Convert Categorical Ranges to Numeric**    
4. **Encode Categorical Variables**
    - We use one-hot encoding to convert categorical variables into numerical format.
5. **Encode 'Health rating'**
    - We manually map 'Health rating' categories to numerical values.
6. **Include 'Age' as Numeric Feature**
    - Add the 'Age' column back to the encoded DataFrame.
7. **Check for NaN or Infinite Values Again**
    - Replace any remaining `NaN` values with the median.
8. **Standardize the Data**
    - Normalize the data to have a mean of 0 and a standard deviation of 1.
9. **Find the Optimal Number of Clusters Using the Elbow Method**
    - Plot the sum of squared distances (inertia) to determine the optimal number of clusters.
10. **Apply KMeans Clustering**
    - Using the optimal number of clusters found from the Elbow method (assumed 3 in this case), apply KMeans clustering.
11. **Visualize the Clusters**
    - Generate a pairplot to visualize the clusters with respect to 'Health rating' and 'Age'.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler



# Replace infinite values with NaN
df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Handle missing values
# Fill NaN values for numeric columns with the median of each column
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].median())

# Convert categorical ranges to numeric
def convert_ranges_to_numeric(value):
    if isinstance(value, str):
        # Handle specific known ranges, e.g., '21-25'
        if '-' in value:
            start, end = value.split('-')
            return (int(start) + int(end)) / 2  # Use the midpoint of the range
        else:
            try:
                return float(value)
            except ValueError:
                return np.nan  # Handle cases where conversion is not possible
    return value

# Apply the conversion function to relevant columns
range_columns = ['Age']  # Example column that might have ranges, adjust as needed
for col in range_columns:
    df[col] = df[col].apply(convert_ranges_to_numeric)

# Encode categorical variables
categorical_columns = df.select_dtypes(include=['object']).columns
df_encoded = pd.get_dummies(df, columns=categorical_columns)

# Encode 'Health rating' numerically
health_rating_mapping = {
    'Excellent': 4,
    'Good': 3,
    'Fair': 2,
    'Poor': 1
}
df_encoded['Health rating'] = df['Health rating'].map(health_rating_mapping)

# Include 'Health rating' in encoded data
df_encoded = pd.get_dummies(df, columns=[col for col in categorical_columns if col != 'Health rating'])
df_encoded['Health rating'] = df['Health rating'].map(health_rating_mapping)

# Add 'Age' as a numeric feature to encoded data
df_encoded['Age'] = df['Age']

# Check for NaN or infinite values again
df_encoded.replace([np.inf, -np.inf], np.nan, inplace=True)
df_encoded.fillna(df_encoded.median(), inplace=True)  # Replace NaN with the median of each column

# Standardizing the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_encoded)

# Finding the optimal number of clusters using the elbow method
inertia = []
for n in range(1, 11):
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Increase the size of the Elbow Method plot
plt.figure(figsize=(10, 6))  # Adjust width and height as needed
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()

# Applying KMeans clustering with the optimal number of clusters
optimal_clusters = 3  # Assume 3 clusters for this example
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
df_encoded['Cluster'] = kmeans.fit_predict(df_scaled)

# Generate pairplot with 'Health rating' and 'Age'
plt.figure(figsize=(12, 10))  # Increase size of the pairplot
sns.pairplot(df_encoded, hue='Cluster', vars=['Health rating', 'Age'])
plt.suptitle('Clustering of Students with Health Rating and Age', y=1.02)  # Adjust title position
plt.show()
```

##  Data  Preparation

We start by defining a function to identify columns with null values and remove the 'Names' column if it exists. Null values in the identified columns are then replaced with 'unknown'. The 'Health rating' column is standardized to ensure it contains expected values, and rows with 'unknown' health ratings are removed. Finally, we print the shape of the cleaned DataFrame and display its contents to verify the changes.


```python
# Define seek_null function
def seek_null(dataframe):
    return dataframe.columns[dataframe.isnull().any()].tolist()

# Delete the 'Names' column if it exists
if 'Names' in df.columns:
    del df['Names']

# Find columns with null values
seeked = seek_null(df)

# Fill null values with 'unknown'
for col in seeked:
    df[col] = df[col].fillna('unknown')

# Ensure 'Health rating' has expected values
df['Health rating'] = df['Health rating'].replace({
    'Excellent;Good': 'Excellent',
    'Good;Fair': 'Fair',
    'Excellent;Good;Fair;Poor': 'Excellent'
})

# Remove rows where 'Health rating' is 'unknown'
df = df[df['Health rating'] != 'unknown']

# Display the shape of the DataFrame
print(df.shape)

# Display the updated DataFrame
print(df)
```
